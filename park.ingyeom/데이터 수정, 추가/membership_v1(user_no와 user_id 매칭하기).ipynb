```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
membership = pd.read_excel("../Dataset/Membership.xlsx")
usermapping = pd.read_excel("../Dataset/User_Mapping.xlsx")
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

1. Membership을 copy해 membership_v1 생성

```python
membership_v1 = membership.copy()
```

2. User_Mapping의 uid와 USER_ID, 그리고 Membership의 user_no를 merge
3. 111개의 중복되는 UID도 삭제.

```python
membership_v1 = membership_v1.merge(
	usermapping[["uid", "USER_ID"]].drop_duplicates(subset=["uid"]),
	left_on="user_no",
	right_on="uid",
	how="left"
)
```

4. 찌꺼기 (uid와 user_no)는 버리고, USER_ID만을 PK로 남긴 채 열 순서 조정

```python
membership_v1 = membership_v1[["USER_ID", "product_cd", "amount", "billing_method", "concurrent_streams", "promotion_yn", "is_churn_prevented", "repurchase", "payment_device", "is_user_verified", "gender", "age", "reg_date", "reg_hour", "end_date"]]
```

5. Excel로 저장

```python
membership_v1.to_excel("../데이터 수정, 추가/membership_v1.xlsx", index=False)
```
