```python
from pypdf import PdfReader, PdfWriter
from copy import deepcopy
```

```python
input_pdf = r"..\..\..\_docs\스터디\1_파이썬 라이브러리를 활용한 머신러닝 (사이킷런 핵심 개발자가 쓴 머신러닝과 데이터 과학 실무서).pdf"
output_pdf = r"..\..\..\_docs\스터디\temp.pdf"
```

```python
reader = PdfReader(input_pdf)
writer = PdfWriter()

for page in reader.pages:
    width = float(page.mediabox.width)
    height = float(page.mediabox.height)

    mid = width / 2

    # 왼쪽 페이지
    left_page = deepcopy(page)
    left_page.mediabox.lower_left = (0, 0)
    left_page.mediabox.upper_right = (mid, height)

    # 오른쪽 페이지
    right_page = deepcopy(page)
    right_page.mediabox.lower_left = (mid, 0)
    right_page.mediabox.upper_right = (width, height)

    # 순서대로 추가
    writer.add_page(left_page)
    writer.add_page(right_page)

with open(output_pdf, "wb") as f:
    writer.write(f)

print("완료:", output_pdf)
```
