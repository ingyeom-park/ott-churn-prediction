```python
import requests
import csv
from datetime import date, timedelta
```

```python
API_KEY = "470e9ac05da21ed137381f0b878e4ea7"
START_DATE = date(2021, 3, 1)
END_DATE = date(2021, 4, 7)
OUTPUT_CSV = "kobis_boxoffice_20210301_20210407.csv"
URL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json"
```

```python
rows = []
current = START_DATE

while current <= END_DATE:
    target_dt = current.strftime("%Y%m%d")
    params = {"key": API_KEY, "targetDt": target_dt}

    res = requests.get(URL, params=params)
    result = res.json()["boxOfficeResult"]
    boxoffice_type = result["boxofficeType"]
    show_range = result["showRange"]

    for item in result["dailyBoxOfficeList"]:
        rows.append({
            "boxofficeType": boxoffice_type,
            "showRange": show_range,
            "rnum": item.get("rnum", ""),
            "rank": item.get("rank", ""),
            "rankInten": item.get("rankInten", ""),
            "rankOldAndNew": item.get("rankOldAndNew", ""),
            "movieCd": item.get("movieCd", ""),
            "movieNm": item.get("movieNm", ""),
            "openDt": item.get("openDt", ""),
            "salesAmt": item.get("salesAmt", "0"),
            "salesShare": item.get("salesShare", "0"),
            "salesInten": item.get("salesInten", "0"),
            "salesChange": item.get("salesChange", "0"),
            "salesAcc": item.get("salesAcc", "0"),
            "audiCnt": item.get("audiCnt", "0"),
            "audiInten": item.get("audiInten", "0"),
            "audiChange": item.get("audiChange", "0"),
            "audiAcc": item.get("audiAcc", "0"),
            "scrnCnt": item.get("scrnCnt", "0"),
            "showCnt": item.get("showCnt", "0"),
        })

    print(f"{target_dt} 완료")
    current += timedelta(days=1)

with open(OUTPUT_CSV, "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.DictWriter(f, fieldnames=rows[0].keys())
    writer.writeheader()
    writer.writerows(rows)

print(f"저장 완료: {OUTPUT_CSV}")
```
