```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
membership = pd.read_excel("../Dataset/Membership.xlsx")
viewhistory = pd.read_excel("../Dataset/View_History.xlsx")
usermapping = pd.read_excel("../Dataset/User_Mapping.xlsx")
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

>구독한 뒤 1개라도 시청한 사람 VS 아무것도 안 본 사람

```python
# 단위기간의 구독 유저 수 (19877명)
entire_N = len(usermapping)

# Active한 유저 수 (14892명)
active_N = len(viewhistory["USER_ID"].unique())

# Non-Active한 유저 수 
non_active_N = entire_N-active_N

plt.pie(
    [active_N, non_active_N],
    labels=[f"Active\n({active_N}명)", f"Non-Active\n({non_active_N}명)"],
    autopct="%1.1f%%"
)
plt.legend(title="User Type", loc="lower left")
plt.title("Active vs Non-Active Users")
plt.show()
```
